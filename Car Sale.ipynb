###### (ii) Checking Multicollinearity

```python
data_cleaned.columns.values
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
variables = data_cleaned[['Mileage','Year','EngineV']]
VIF = pd.DataFrame()
VIF['VIF'] = [vif(variables.values, i) for i in range(variables.shape[1])]
VIF['Features'] = variables.columns
VIF
```

Since the VIF of "Year" is greater that 5, we will drop the feature "Year".

```python
data_no_mc = data_cleaned.drop(['Year'], axis = 1)
data_no_mc.head()
```

# CREATE DUMMY VARIABLES

```python
data_with_dummies = pd.get_dummies(data_no_mc,drop_first = True)
data_with_dummies.head()
```

```python
data_with_dummies.columns.values
```

```python
cols = ['Log Price', 'Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed.head()
```

# LINEAR REGRESSION

###### (i) Declare the inputs and targets

```python
targets = data_preprocessed['Log Price']
inputs = data_preprocessed.drop(['Log Price'], axis = 1)
```

###### (ii)Scale the data

```python
from sklearn.preprocessing import StandardScaler as SS
```

```python
scaler = SS()
```

```python
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

###### (iii) Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size = 0.2, random_state = 365)
```

###### (iv)Create the regression

```python
reg = LR()
```

```python
reg.fit(x_train, y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat)
plt.xlabel('Targets(y_train)', size = 18)
plt.ylabel('Predictions(y_hat)', size = 18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
sns.distplot(y_train-y_hat)
plt.title('Residuals PDF', size = 18)
```

```python
reg.score(x_train, y_train)
```

###### (v) Find the weights and the bias

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns = ['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

# TESTING

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test, y_hat_test, alpha = 0.2)
plt.xlabel('Targets(y_test)', size = 18)
plt.ylabel('Predictors(y_hat_test)', size = 18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test), columns = ['Prediction'])
df_pf.head()
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
y_test
```

```python
y_test = y_test.reset_index(drop = True)
y_test.head()
```

```python
y_test.head()
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
df_pf['Residual'] = df_pf['Target'] - df_pf['Prediction']
```

```python
df_pf['Difference%'] = np.absolute(df_pf['Residual']/df_pf['Target']*100)
df_pf.describe()
```
